```python
import os
import pandas as pd

accident_dir = 'data\\accident'
non_accident_dir = 'data\\non-accident'

images = []
labels = []

# Function to process a folder and assign the folder name as the label
def process_folder(folder_path, label):
    files = os.listdir(folder_path)
    for filename in files:
        # full_path = os.path.join(folder_path, filename)
        images.append(filename)
        labels.append(label)

process_folder(accident_dir, os.path.basename(accident_dir))
process_folder(non_accident_dir, os.path.basename(non_accident_dir))

df = pd.DataFrame({
    'image': images,
    'label': labels
})

# df.sort_values(['label', 'image'], ascending=[True, True])
df.to_csv('accident_labels.csv', index=False)
df
```
